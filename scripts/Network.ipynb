```python
from whatnet.data.datasets.mnist import read_data_sets
mnist = read_data_sets('./data/')

from whatnet.data.converter import ConverterBase
converter = ConverterBase(28 * 28)
```

```python
from whatnet.network import Network

net = Network(converter)
input_layer = net.create_neuron(int((28 * 28 + 7) / 8))
output_layer = net.create_neuron(10)
net.create_synapse(input_layer, output_layer, conn_spec="all_to_all")
net.mark_input(input_layer)
net.mark_output(output_layer)
```

```python
#for image, target in zip(mnist.train.data[:100], mnist.train.target[:100]):
    #logging.info(target)
    #net.train(image, target)
image, target = mnist.train.data[1], mnist.train.target[1]
net.train(image, target)
image, target = mnist.train.data[0], mnist.train.target[0]
net.train(image, target)
print(target)
```
